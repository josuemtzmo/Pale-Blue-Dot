### Load data

```python
# Daily max temperature
temp_max = xr.open_mfdataset(output_path + "tasmax_day_ACCESS-CM*.nc",chunks={"time":365,"lat":300,'lon':480})
# Daily temperature
temp = xr.open_mfdataset(output_path + "tas_day_ACCESS-CM2_*.nc",chunks={"time":365,"lat":300,'lon':480})
# Daily humidity
huss = xr.open_mfdataset(output_path + "huss_day_ACCESS-CM2_*.nc",chunks={"time":365,"lat":300,'lon':480})
```

```python
# temp_max
```

```python
monthly_temp_max = temp_max.resample(time='1M').mean()
monthly_temp = temp.resample(time='1M').mean()
monthly_huss = huss.resample(time='1M').mean()
```

```python
years = np.unique(monthly_temp_max['time.year'])
years[129:]
```

```python
for year in years[129:]:
    print(year)
    monthly_temp_max.sel(time="{0}".format(year)).to_netcdf(output_path+"monthly/tasmax_day_ACCESS-CM_{0}.nc".format(year))
    monthly_temp.sel(time="{0}".format(year)).to_netcdf(output_path+"monthly/tas_day_ACCESS-CM_{0}.nc".format(year))
    monthly_huss.sel(time="{0}".format(year)).to_netcdf(output_path+"monthly/huss_day_ACCESS-CM_{0}.nc".format(year))
```
